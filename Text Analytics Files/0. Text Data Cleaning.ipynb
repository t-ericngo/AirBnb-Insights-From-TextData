## Data Cleaning

### 1. Bangkok

```python
import pandas as pd
df1 = pd.read_csv("BangkokListing.csv")
df2 = pd.read_csv("BangkokReviews.csv")
dfBK = pd.merge(df1, df2, on='listing_id', how='left')
dfBK.head()
```

#### Language Detection & Removal of Non-English Comments

```python
#!pip install langdetect
from langdetect import detect

# Create a function to detect the language of a comment
def detect_language(text):
    try:
        lang = detect(text)
        return lang == 'en'  # Return True if language is English, False otherwise
    except:
        return False  # Return False if language detection fails

# Apply the function to detect the language of each comment and create a mask
mask = dfBK['comments'].apply(detect_language)

mask
```

```python
# Filter the DataFrame to keep only the rows where the language is English
dfBK_eng = dfBK[mask]

dfBK_eng.reset_index(drop=True, inplace=True)

dfBK_eng.head(10)
```

#### Filter Time Frame

```python
#filter data 2016 to 2023
from datetime import datetime

dfBK_eng.dropna(subset=["date"], inplace = True)
dfBK_eng["timestamp"] = pd.to_datetime(dfBK_eng["date"], format = "%m/%d/%y")

# Sort rows by listing id & then by time stamp
dfBK_eng.sort_values(by = ["listing_id", "timestamp"], ascending = True, inplace = True)
dfBK_eng.reset_index(drop = True, inplace = True)

start_date = pd.to_datetime("01/01/16")
end_date = pd.to_datetime("12/31/23")

dfBK_eng = dfBK_eng[(dfBK_eng["timestamp"] >= start_date) & (dfBK_eng["timestamp"] <= end_date)]

# Sort rows by listing id & then by time stamp - again
dfBK_eng.sort_values(by = ["listing_id", "timestamp"], ascending = True, inplace = True)
dfBK_eng.reset_index(drop = True, inplace = True)
```

```python
### create a variable for host's experience (total months of hosting)
import numpy as np
dfBK_eng['host_since1'] = pd.to_datetime(dfBK_eng['host_since'], format = "%m/%d/%y")

dfBK_eng['host_exp_month'] = (dfBK_eng['timestamp'].dt.year - dfBK_eng['host_since1'].dt.year) * 12 + (dfBK_eng['timestamp'].dt.month - dfBK_eng['host_since1'].dt.month)

dfBK_eng.head(10)
```

```python
## Drop variables
dfBK_eng = dfBK_eng.drop(columns=['host_location','host_verifications','host_has_profile_pic','host_neighbourhood','availability_90','minimum_nights','review_scores_checkin','instant_bookable','reviews_per_month'])
```

#### Cleaning Text Data

```python
## Cleaning reviews/comments
import re
dfBK_eng["comments"] = dfBK_eng["comments"].str.lower()
dfBK_eng["comments"] = [str(x).replace("n't "," not ") for x in dfBK_eng.comments]  # replace "n't " with " not "
dfBK_eng["comments"] = [str(x).replace(" br ","") for x in dfBK_eng.comments] 
dfBK_eng["comments"] = [str(x).replace("br ","") for x in dfBK_eng.comments] 
dfBK_eng["comments"] = [str(x).replace(" br","") for x in dfBK_eng.comments] 
dfBK_eng["comments"] = [str(x).replace("br","") for x in dfBK_eng.comments] 
dfBK_eng["comments"] = [re.sub('[^a-zA-Z0-9]', ' ', x) for x in dfBK_eng.comments]
dfBK_eng["comments"].head(10)
```

```python
## Cleaning neighborhood overview
dfBK_eng["neighborhood_overview"] = dfBK_eng["neighborhood_overview"].str.lower()
dfBK_eng["neighborhood_overview"] = [str(x).replace("n't "," not ") for x in dfBK_eng.neighborhood_overview]  # replace "n't " with " not "
dfBK_eng["neighborhood_overview"] = [str(x).replace(" br ","") for x in dfBK_eng.neighborhood_overview] 
dfBK_eng["neighborhood_overview"] = [str(x).replace("br ","") for x in dfBK_eng.neighborhood_overview] 
dfBK_eng["neighborhood_overview"] = [str(x).replace(" br","") for x in dfBK_eng.neighborhood_overview] 
dfBK_eng["neighborhood_overview"] = [str(x).replace("br","") for x in dfBK_eng.neighborhood_overview] 
dfBK_eng["neighborhood_overview"] = [re.sub('[^a-zA-Z0-9]', ' ', x) for x in dfBK_eng.neighborhood_overview]
dfBK_eng["neighborhood_overview"].head(5)
```

```python
## Cleaning host about
dfBK_eng["host_about"] = dfBK_eng["host_about"].str.lower()
dfBK_eng["host_about"] = [str(x).replace("n't "," not ") for x in dfBK_eng.host_about]  # replace "n't " with " not "
dfBK_eng["host_about"] = [str(x).replace(" br ","") for x in dfBK_eng.host_about] 
dfBK_eng["host_about"] = [str(x).replace("br ","") for x in dfBK_eng.host_about] 
dfBK_eng["host_about"] = [str(x).replace(" br","") for x in dfBK_eng.host_about] 
dfBK_eng["host_about"] = [str(x).replace("br","") for x in dfBK_eng.host_about] 
dfBK_eng["host_about"] = [re.sub('[^a-zA-Z0-9]', ' ', x) for x in dfBK_eng.host_about]
dfBK_eng["host_about"].head(10)
```

#### Export Clean Data

```python
dfBK_eng.to_csv('BKclean_data.csv', index=False)
```

### 2. San Francisco

```python
import pandas as pd
df3 = pd.read_csv("SFlistings.csv")
df4 = pd.read_csv("SFreviews.csv")
dfSF = pd.merge(df3, df4, on='listing_id', how='left')
dfSF
```

#### Language Detection & Removal of Non-English Comments

```python
# !pip install langdetect
from langdetect import detect

# Create a function to detect the language of a comment
def detect_language(text):
    try:
        lang = detect(text)
        return lang == 'en'  # Return True if language is English, False otherwise
    except:
        return False  # Return False if language detection fails

# Apply the function to detect the language of each comment and create a mask
mask = dfSF['comments'].apply(detect_language)

# Filter the DataFrame to keep only the rows where the language is English
dfSF = dfSF[mask]

dfSF.reset_index(drop=True, inplace=True)

dfSF
```

#### Filter Time Frame

```python
#filter data 2016 to 2023
dfSF.loc[:, 'timestamp'] = pd.to_datetime(dfSF['date'], format = "%m/%d/%y")

start_date = pd.to_datetime("2016-01-01")
end_date = pd.to_datetime("2023-12-31")
dfSF = dfSF[(dfSF["timestamp"] >= start_date) & (dfSF["timestamp"] <= end_date)]

dfSF
```

```python
#convert host_since to host experience in months
dfSF.loc[:,'host_since1'] = pd.to_datetime(dfSF['host_since'], format = "%m/%d/%y")

dfSF.loc[:,'host_exp_month'] = (dfSF['timestamp'].dt.year - dfSF['host_since1'].dt.year) * 12 + (dfSF['timestamp'].dt.month - dfSF['host_since1'].dt.month)

dfSF
```

```python
#drop unnecessary columns
dfSF = dfSF.drop(columns=['host_neighbourhood','host_since1','host_since','host_verifications'])
dfSF
```
